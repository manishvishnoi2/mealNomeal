```python
import warnings
import pandas as pd
import numpy as np

from tsfresh import extract_features
from tsfresh.utilities.dataframe_functions import impute
```

```python
result = pd.DataFrame()
for x in range(5):
    d = pd.read_csv('data/mealData'+str(x+1)+'.csv', header = None,error_bad_lines=False)
    d['y']= 1
    result = pd.concat([result,d])
```

```python
for x in range(5):
    d = pd.read_csv('data/Nomeal'+str(x+1)+'.csv', header = None,error_bad_lines=False)
    d['y']= 0
    result = pd.concat([result,d])
```

```python
result.tail()
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan,strategy='mean')
imp_mean.fit(result)
result = imp_mean.transform(result)
```

```python
result = pd.DataFrame(result)
```

```python
result.tail()
```

```python
columns = list(result.columns)
columns.pop()
columns.append('target')
result.columns = columns
```

```python
y = result.target
result.drop( 'target', axis = 1, inplace = True )
```

```python
result.head()
```

```python
d = result.stack()
d.index.rename([ 'id', 'time' ], inplace = True )
d = d.reset_index()
```

```python
d.shape
```

```python
with warnings.catch_warnings():
        warnings.simplefilter( "ignore" )
        f = extract_features( d, column_id = "id", column_sort = "time" )
```

```python
f.shape
```

```python
impute(f)
assert f.isnull().sum().sum() == 0

f['y'] = y
f.to_csv( "data/features_file.csv", index = None )
```
