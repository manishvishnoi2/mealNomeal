```python
import pandas as pd
from tsfresh import extract_features, select_features
from sklearn.model_selection import train_test_split
```

```python
features = pd.read_csv("data/features_file.csv")
```

```python
features['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features.loc[:, features.columns != 'y'], features['y'], test_size=0.33, random_state=42)
```

```python
train_features_selected = select_features( X_train, y_train, fdr_level = 0.0005 )
```

```python
print ("selected {} features.".format( len( train_features_selected.columns )))
```

```python
train_features_selected
```

```python
train = train_features_selected.copy()
train['y'] = y_train

test = X_test[train_features_selected.columns].copy()
test['y'] = y_test
```

```python
test.shape
```

```python
train.to_csv( "data/train_file.csv", index = None )
test.to_csv( "data/test_file.csv", index = None )
```
